```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
# It's better to preprocess the raw data with Excel because the original output format sucks.
data = pd.read_csv("emg.txt", sep='\t')
```

```python
data["c1"].plot()
```

```python
data["c2"].plot()
```

```python
plt.xlabel(r"$time [1/200sec]$")
plt.ylabel(r"$voltage[\mu V]$")
data["c1"][:410000].plot(alpha=0.7, label="biceps brachii muscle")
data["c2"][:410000].plot(alpha=0.7, label="triceps brachii muscle")
plt.legend(loc=3)
plt.savefig("bt.png", dpi=400)
```
