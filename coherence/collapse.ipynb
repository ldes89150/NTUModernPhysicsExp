```python
import numpy as np
import scipy as sp
from scipy.misc import imread
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
outputName = "bulb_double_polarizer"+".csv"
background=[
    "bulb_double_polarizer_bg_0_150602-145956.bmp"
]
data = [
    "bulb_double_0_150602-145657.bmp",
    "bulb_double_0_150602-145658.bmp",
    "bulb_double_0_150602-145700.bmp",
    "bulb_double_0_150602-145701.bmp",
    "bulb_double_0_150602-145702.bmp"
]
showImg = True
```

```python
bgImg= [imread(i) for i in background]
dataImg = [imread(i) for i in data]
if showImg:
    for i in bgImg:
        plt.imshow(i)
        plt.show()
    for i in dataImg:
        plt.imshow(i)
        plt.show()
```

```python
b = np.average(bgImg,axis=0)
d = np.average(dataImg,axis=0)
```

```python
signal = d-b
plt.imshow(signal)
```

```python
collapse = np.sum(d,axis=0)
plt.plot(collapse)
```

```python
collapse.tofile(outputName,sep=',')
```
